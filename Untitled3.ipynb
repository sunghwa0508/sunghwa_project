```python
def outliers_iqr(data):
    q1, q3 =np.percentile(data, [25,75])
    #넘파이의 값을 퍼센트로 표시
    iqr = q3 - q1
    lower_bound = q1 - (iqr*1.5)
    upper_bound = q3 + (iqr*1.5)

    return np.where((data>upper_bound)|(data<lower_bound))
```

```python
#이상치 확인
balance_outlier_index = outliers_iqr(data['balance'])[0]
balance_outlier_index
```

```python
# deposit 평균값을 base value로 합니다.
explainer.expected_value
```

```python
#예측함수를 정의합니다. 

def predict(age,balance,day,duration,campaign,pdays,previous,default_no,default_yes):

    # 함수 내에서 예측에 사용될 input을 만듭니다
    df = pd.DataFrame(
        data=[[age,balance,day,duration,campaign,pdays,previous,default_no,default_yes]], 
        columns=['age','balance','day','duration','campaign','pdays','previous','default_no','default_yes']
    )

    # 예측
    pred = model.predict(ddf)[0]

    # Shap value를 계산합니다
    explainer = shap.TreeExplainer(model)
    shap_values = explainer.shap_values(ddf)

    # Shap value, 특성이름, 특성값을 가지는 Series를 만듭니다
    feature_names = ddf.columns
    feature_values = ddf.values[0]
    shaps = pd.Series(shap_values[0], zip(feature_names, feature_values))

    # 결과를 프린트 합니다.
    result = f'평균 deposit: ${explainer.expected_value[0]:,.0f} \n'
    result += f'예측 deposit: ${pred:,.0f}. \n'
    result += shaps.to_string()
    print(result)


    # SHAP Force Plot
    shap.initjs()
    return shap.force_plot(
        base_value=explainer.expected_value, 
        shap_values=shap_values, 
        features=df
    )
```
