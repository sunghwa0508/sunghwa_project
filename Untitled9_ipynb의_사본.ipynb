```python
import json
import numpy as np
import pandas as pd

file = open('trending.json', encoding="utf8")

# Load data as JSON
raw_data = json.load(file)

# Close the original file
file.close()

# Select only the list with the video data
trending_videos_list = raw_data['collector']

# Example of a video object
print(json.dumps(trending_videos_list[15], indent=4, sort_keys=True))
```

```python
# Create a DataFrame of the data
df_tiktok_dataset = pd.DataFrame(trending_videos_list)

# Let's expand the hashtag cell containing lists to multiple rows
df_tiktok_dataset = df_tiktok_dataset.explode('hashtags').explode('mentions')
```

```python
def object_to_columns(dfRow, **kwargs):
    '''Function to expand cells containing dictionaries, to columns'''
    for column, prefix in kwargs.items():
        if isinstance(dfRow[column], dict):
            for key, value in dfRow[column].items():
                columnName = '{}.{}'.format(prefix, key)
                dfRow[columnName] = value
    return dfRow

# Expand certain cells containing dictionaries to columns
df_tiktok_dataset = df_tiktok_dataset.apply(object_to_columns, 
                            authorMeta='authorMeta',  
                            musicMeta='musicMeta',
                            covers='cover',
                            videoMeta='videoMeta',
                            hashtags='hashtag', axis = 1)

# Remove the original columns containing the dictionaries
df_tiktok_dataset = df_tiktok_dataset.drop(['authorMeta','musicMeta','covers','videoMeta','hashtags'], axis = 1)
df_tiktok_dataset
```

```python
import cv2
from google.colab.patches import cv2_imshow


# 영상 검출기
def videoDetector(cam,cascade,age_net,gender_net,MODEL_MEAN_VALUES,age_list,gender_list):

    while True:

        # 캡처 이미지 불러오기
        ret,img = cam.read()
        # 영상 압축
        try:
            img = cv2.resize(img,dsize=None,fx=1.0,fy=1.0)
        except: break
        # 그레이 스케일 변환
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
        # cascade 얼굴 탐지 알고리즘 
        results = cascade.detectMultiScale(gray,            # 입력 이미지
                                           scaleFactor= 1.1,# 이미지 피라미드 스케일 factor
                                           minNeighbors=5,  # 인접 객체 최소 거리 픽셀
                                           minSize=(20,20)  # 탐지 객체 최소 크기
                                           )
        

        for box in results:
            x, y, w, h = box
            face = img[int(y):int(y+h),int(x):int(x+h)].copy()
            blob = cv2.dnn.blobFromImage(face, 1, (227, 227), MODEL_MEAN_VALUES, swapRB=False)

            # gender detection
            gender_net.setInput(blob)
            gender_preds = gender_net.forward()
            gender = gender_preds.argmax()
            # Predict age
            age_net.setInput(blob)
            age_preds = age_net.forward()
            age = age_preds.argmax()
            
            info = gender_list[gender] +' '+ age_list[age]

            cv2.rectangle(img, (x,y), (x+w, y+h), (255,255,255), thickness=2)
            cv2.putText(img,info,(x,y-15),0, 0.5, (0, 255, 0), 1)


         # 영상 출력
        cv2_imshow(img)
        
        if cv2.waitKey(1) > 0: 
  
            break


# 얼굴 탐지 모델 가중치
cascade_filename = '/content/drive/My Drive/haarcascade_frontalface_alt.xml'
# 모델 불러오기
cascade = cv2.CascadeClassifier(cascade_filename)


MODEL_MEAN_VALUES = (78.4263377603, 87.7689143744, 114.895847746)

age_net = cv2.dnn.readNetFromCaffe(
	'/content/drive/My Drive/deploy_age.prototxt',
	'/content/drive/My Drive/age_net.caffemodel')

gender_net = cv2.dnn.readNetFromCaffe(
	'/content/drive/My Drive/deploy_gender.prototxt',
	'/content/drive/My Drive/gender_net.caffemodel')

age_list = ['(0 ~ 2)','(4 ~ 6)','(8 ~ 12)','(15 ~ 20)',
            '(25 ~ 32)','(38 ~ 43)','(48 ~ 53)','(60 ~ 100)']
gender_list = ['Male', 'Female']

# 영상 파일 
cam = cv2.VideoCapture('videos/6875373441432816898.mp4')

# 영상 탐지기
videoDetector(cam,cascade,age_net,gender_net,MODEL_MEAN_VALUES,age_list,gender_list)
```

#### 텍스트 유사도를 이용한 추천 시스템

```python
dff = df1[['id','diggCount'	,'authorMeta.id',	'authorMeta.nickName',	'musicMeta.musicAuthor',	'musicMeta.musicName'	,'playCount'	,'shareCount',	'text', 'webVideoUrl']]
dff
```

```python
#중복 id 첫번째만 남기고 제거
dff2 = dff.drop_duplicates(['id'], keep = 'first')
dff2
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

tfidf = TfidfVectorizer(stop_words='english')
# overview에 대해서 tf-idf 수행
tfidf_matrix = tfidf.fit_transform(dff2['text'])
print(tfidf_matrix.shape)
```

```python
#코사인 유사도

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

indices = pd.Series(dff2.index, index=dff2['authorMeta.nickName']).drop_duplicates()
print(indices.head())
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    # 선택한  타이틀로부터 해당되는 인덱스를 받아옵니다.
    idx = indices[title]

    # 모든 영상에 대해서 해당 영상과의 유사도를 구합니다.
    sim_scores = list(enumerate(cosine_sim[idx]))

    # 유사도에 따라 닉네임을 정렬합니다.
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # 가장 유사한 10개닉네임을 받아옵니다.
    sim_scores = sim_scores[1:11]

    # 가장 유사한 10개의 인덱스를 받아옵니다.
    movie_indices = [i[0] for i in sim_scores]

    # 가장 유사한 10개의 틱톡 닉네임을 리턴합니다.
    return print(dff2['authorMeta.nickName'].iloc[movie_indices] + ' ' + dff2['webVideoUrl'].iloc[movie_indices])

get_recommendations('JackJacko')
```
